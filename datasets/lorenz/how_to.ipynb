```python
import pickle
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

```python
with open("./lorenz_dataset.pickle","rb") as infile:
    dataset = pickle.load(infile)
```

```python
print("Sample condtional window from each condition cluster")
plt.rcParams["patch.force_edgecolor"] = True
matplotlib.rcParams.update({'font.size': 14})
plt.figure(figsize=(15,20))
plt.title("test")
for i in range(5):
    ax = plt.subplot('32'+str(i+1))
    for ts in dataset["x_train"]["condition_{}".format(i)][:200]:
        plt.plot(ts,alpha = 0.1,color="coral")
    ax.set_title("\n condition cluster {}".format(i))

    ax.set_ylim(-10,10)

plt.show()
```

```python
print("The distribution of X(t+1) in train set")
plt.rcParams["patch.force_edgecolor"] = True
matplotlib.rcParams.update({'font.size': 14})
plt.figure(figsize=(15,20))
plt.title("test")
for i in range(6):
    ax = plt.subplot('32'+str(i+1))
    if i !=5:
        plt.hist(dataset["y_train"]["condition_{}".format(i)], bins=80, density=True, range=(-11, 11),color="coral",edgecolor="r")
        ax.set_title("\n condition cluster {}".format(i))
    else:
        plt.hist(np.concatenate(list(dataset["y_train"].values())), bins=80, density=True, range=(-11, 11),color="coral",edgecolor="r")
        ax.set_title("\n Entire train set")
    ax.set_xlabel(r'$x_{t+1}$')
    ax.set_ylabel(r'$\rho(x_{t+1})$')
    ax.set_ylim(0,0.33)

plt.show()
```

```python
print("The distribution of X(t+1) in validation set")
plt.rcParams["patch.force_edgecolor"] = True
matplotlib.rcParams.update({'font.size': 14})
plt.figure(figsize=(15,20))
plt.title("test")
for i in range(6):
    ax = plt.subplot('32'+str(i+1))
    if i !=5:
        plt.hist(dataset["y_val"]["condition_{}".format(i)], bins=80, density=True, range=(-11, 11),color="coral",edgecolor="r")
        ax.set_title("\n condition cluster {}".format(i))
    else:
        plt.hist(np.concatenate(list(dataset["y_val"].values())), bins=80, density=True, range=(-11, 11),color="coral",edgecolor="r")
        ax.set_title("\n Entire val set")
    ax.set_xlabel(r'$x_{t+1}$')
    ax.set_ylabel(r'$\rho(x_{t+1})$')
    ax.set_ylim(0,0.33)

plt.show()
```

```python
print("The distribution of X(t+1) in test set")
plt.rcParams["patch.force_edgecolor"] = True
matplotlib.rcParams.update({'font.size': 14})
plt.figure(figsize=(15,20))
plt.title("test")
for i in range(6):
    ax = plt.subplot('32'+str(i+1))
    if i !=5:
        plt.hist(dataset["y_test"]["condition_{}".format(i)], bins=80, density=True, range=(-11, 11),color="coral",edgecolor="r")
        ax.set_title("\n condition cluster {}".format(i))
    else:
        plt.hist(np.concatenate(list(dataset["y_test"].values())), bins=80, density=True, range=(-11, 11),color="coral",edgecolor="r")
        ax.set_title("\n Entire test set")
    ax.set_xlabel(r'$x_{t+1}$')
    ax.set_ylabel(r'$\rho(x_{t+1})$')
    ax.set_ylim(0,0.33)

plt.show()
```
